# 2.2 Remove all the stopwords from all three speeches



```python
Roosevelt = inaugural.raw('1941-Roosevelt.txt')
Kennedy = inaugural.raw('1961-Kennedy.txt')
Nixon = inaugural.raw('1973-Nixon.txt')
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
print(stopwords.words('english'))
len(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize
import re
from nltk.stem.snowball import SnowballStemmer
```

```python
import string
stopwords = stopwords.words('english') +list(string.punctuation)
stemmer = SnowballStemmer('english')
```

```python
def Tokenizer(text):
  temp = [w for w in word_tokenize(text)]
  tokens = []
  for t in temp:
    if re.search('[a-zA-Z]', t):
      tokens.append(t)
  return_t = [stemmer.stem(x) for x in tokens]
  return return_t
```

```python
Roosevelt_Filter = [w for w in Tokenizer(Roosevelt) if w not in stopwords]
```

```python
Kennedy_Filter = [w for w in Tokenizer(Kennedy) if w not in stopwords]
```

```python
Nixon_Filter = [w for w in Tokenizer(Nixon) if w not in stopwords]
```

```python
print('Before Stemming and Stopwords removal wordcount of Roosevelt : ', len(inaugural.words('1941-Roosevelt.txt')))
print('After Stemming and Stopwords removal wordcount of Roosevelt : ', len(Roosevelt_Filter))
print('\n')
print('Before Stemming and Stopwords removal wordcount of Kennedy : ', len(inaugural.words('1961-Kennedy.txt')))
print('After Stemming and Stopwords removal wordcount of Kennedy : ', len(Kennedy_Filter))
print('\n')
print('Before Stemming and Stopwords removal wordcount of Nixon : ', len(inaugural.words('1973-Nixon.txt')))
print('After Stemming and Stopwords removal wordcount of Nixon : ', len(Nixon_Filter))
```

# 2.3 Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords).

```python
from nltk.probability import FreqDist
```

```python
FreqDist(Roosevelt_Filter).most_common(15)
```

# 2.4 *Word Clouds after cleaning texts*

```python
from wordcloud import WordCloud
```

```python
import matplotlib.pyplot as plt
```

```python
# All words in single statement

wc_r = ' '.join(Roosevelt_Filter)  # Roosevelt Clean speech
wc_k = ' '.join(Kennedy_Filter) # Kennedy Clean speech
wc_n = ' '.join(Nixon_Filter) # Nixon clean speech
```

```python
wc_roosevelt = WordCloud(width = 500, height = 500, background_color='black', min_font_size=12, random_state=2).generate(wc_r)
wc_kennedy = WordCloud(width = 500, height=500,background_color='black', min_font_size=12,random_state=3 ).generate(wc_k)
wc_nixon = WordCloud(width=500, height=500, min_font_size=12, background_color='black',random_state=4).generate(wc_n) 
```

```python
  plt.figure(figsize=(10, 10), facecolor=None)
  plt.imshow(wc_roosevelt)
  plt.axis('off')
  plt.xlabel('Word Cloud')
  plt.tight_layout(pad= 0)
  print ('Word cloud for Roosevelt After cleaning speech')
  plt.show()

```
